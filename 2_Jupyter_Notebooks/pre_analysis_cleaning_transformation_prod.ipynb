**Notes**

- Column names are standarized (first letter of each word in uppercase and spaces are substituted by underscore) except for the columns DF;Route, RLD and RS.
- We can see that the file is pretty complete: the only column with null values is TE-Code.
- Regarding data types, the Approval_Date column should be Datetime instead of Object.
- Transformations: it would be interesting to create two new columns from the Ingredient column: a column called Number_Of_Ingredients to count the number of APIs in the composition of the drug and another boolean column called Combination_Drug to specify whether the drug is composed by just one or multiple APIs. Also it would be interesting to convert all values in DF;Route to uppercase (metadata indicates not all are uppercase) and split the column DF;Route into two.
- Check for unique values in text columns. Maybe we need aditional transformations.
- We should check for duplicates before and after data cleaning and transformation.

```python
#Check for duplicated rows.

print(f"There are {df_prod.duplicated().sum()} duplicates in this dataframe.")
```

```python
#Check for unique values in type text columns

df_prod['Appl_Type'].value_counts()
#reformat the Appl_Type labels and simplify the TE_Code column
```

```python
#We substitute N by Innovator as it stands for NDA (New Drug Application), A by Generic as it stands for ANDA (Abbreviated New Drug Application). Finally every different value or nulls stay the same.
df_prod['Appl_Type'] = df_prod['Appl_Type'].apply(lambda x: 'Innovator' if x=='N' else 'Generic' if x=='A' else 'NaN' if pd.isnull(x) else x)
```

```python
display(df_prod.head())
```

```python
df_prod['Appl_Type'].value_counts()
```

```python
#Fill null values in TE_Code
df_prod['TE_Code'] = df_prod['TE_Code'].fillna('Not Rated')
df_prod.head()
```

```python
print(f'There are {pd.isna(df_prod['TE_Code']).sum()} null values in TE_Code column.')
```

```python
df_prod.info()
```

```python
#Turn all values from DF;Route column into uppercase and then split it in two:
df_prod[['Dosage_Form','Route']] = df_prod['DF;Route'].str.upper().str.split(';', expand=True)
df_prod = df_prod.drop('DF;Route', axis=1)
```

```python
df_prod.head()
```

```python
#Reorder columns
display(df_prod.columns)
```

```python
new_order = ['Ingredient', 'Dosage_Form', 'Route','Trade_Name', 'Applicant', 'Strength', 'Appl_Type',
       'Appl_No', 'Product_No', 'TE_Code', 'Approval_Date', 'RLD', 'RS',
       'Type', 'Applicant_Full_Name']
df_prod = df_prod.reindex(columns= new_order)
df_prod.head()
```

```python
#Now we will change labels for the column Type:
df_prod['Type'].unique()
```

```python
df_prod['Type'] = df_prod['Type'].replace({"DISCN":"Discontinued","RX":"Prescription"})
df_prod.head()
```

```python
#Now we want to convert the column Approval_Date into datetime.
df_prod['Approval_Date'].value_counts() #to check unique values of the column and their count.
```

```python
#We discover there are 5870 values as "Approved Prior to Jan 1, 1982". To be able to convert the column into datetime, we will first replace these values to 1/1/1982.

df_prod['Approval_Date'] = df_prod['Approval_Date'].str.replace("Approved Prior to Jan 1, 1982","1/1/1982")
```

```python
#Now we try to map month abbreviations and replace them by numbers. This will be necessary to do conversion into datetime

month_dict = {
    "Jan": 1,
    "Feb": 2,
    "Mar": 3,
    "Apr": 4,
    "May": 5,
    "Jun": 6,
    "Jul": 7,
    "Aug": 8,
    "Sep": 9,
    "Oct": 10,
    "Nov": 11,
    "Dec": 12
}

def mapping_func(x):
    for key, value in month_dict.items():
        if key in x:
            return x.replace(key, str(value))
    return x  # Return the original value if no match is found

df_prod['Approval_Date'] = df_prod['Approval_Date'].apply(mapping_func)
```

```python
df_prod.head()
```

```python
df_prod['Approval_Date'] = df_prod['Approval_Date'].str.replace(", ", "/")
df_prod['Approval_Date'] = df_prod['Approval_Date'].str.replace(" ","/")
df_prod.head()
```

```python
df_prod['Approval_Date'] = pd.to_datetime(df_prod['Approval_Date']) #conversion into datetime
df_prod.head()
```

```python
df_prod.info()
```

```python
#Now we are going to generate two new columns Number_Of_Ingredients and Combination_Drug. The first column counts the number of APIs in the drug's composition.

df_prod['Number_Of_Ingredients'] = df_prod['Ingredient'].str.count(";") + 1 #When there are no separators (;) there is just one Ingredient. When there is one separator there are two ingredients,...
df_prod['Number_Of_Ingredients'].value_counts()
```

```python
#Now let's create Combination Drug. It will be Yes if Number of Ingredients is more than 1. No if it's equal to 1.

df_prod['Combination_Drug'] = df_prod['Number_Of_Ingredients'].apply(lambda x: "No" if x==1 else "Yes")
df_prod['Combination_Drug'].value_counts()
```

```python
df_prod.head()
```

```python
#Before exporting the df into an excel first we will simplify labels in the TE_Code column and create a new column: if code contains an A, it means the drug is therapeutically equivalent to the reference listed drug (RLD). If it does not contain an A, it is Not therapeutically equivalent.

df_prod['Therapeutically_Equivalent'] = df_prod['TE_Code'].apply(lambda x: "Yes" if "A" in x else x if x=="Not Rated" else "No")
df_prod = df_prod.drop(['TE_Code'], axis = 1)
df_prod.head()
```

```python
df_prod.columns
```

```python
#Finally we decide to remove the column Strength as it won't be necessary for the analysis. Also we will re-check duplicates after all these transformations.

df_prod = df_prod.drop('Strength', axis = 1)
print(f'There are {df_prod.duplicated().sum()} duplicates in the dataframe.')
```

```python
new_order = ['Ingredient', 'Number_Of_Ingredients', 'Combination_Drug', 'Dosage_Form', 'Route', 'Trade_Name', 'Applicant',
       'Appl_Type', 'Appl_No', 'Product_No', 'Approval_Date', 'RLD', 'RS', 'Therapeutically_Equivalent',
       'Type', 'Applicant_Full_Name']

df_prod = df_prod.reindex(columns=new_order)
```

```python
df_prod.rename(columns = {"RLD": "Reference_Listed_Drug", "RS": "Reference_Standard"}, inplace=True) #standarize column names.
```
